- ## 22nd step:
  ### for loops nested


```python
first_names = ["Ethan", "Sophia", "Marcus", "Elena"]
last_names = ["Vance", "Chen", "Rodger", "Kim"]
full_name = []
for a_first_name in first_names:
    for a_last_name in last_names:
        full_name.append(a_first_name + " " + a_last_name)
print(full_name)
```
